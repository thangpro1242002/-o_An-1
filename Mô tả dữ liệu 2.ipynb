```python
import random
import pandas as pd
from time import sleep
import csv
import numpy as np
import datetime 
from datetime import date
from datetime import datetime
import matplotlib 
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
data = pd.read_csv('D:\LamDong1.csv')
data
```

```python
#Tao cot thang luu tru tu cot ngay luu tru
data["Month of stay"] = data["HDate"].apply(lambda x: "".join(re.findall(r'(Jan(?:uary)?|Feb(?:ruary)?|Mar(?:ch)?|Apr(?:il)?|May|Jun(?:e)?|Jul(?:y)?|Aug(?:ust)?|Sep(?:tember)?|Oct(?:ober)?|Nov(?:ember)?|Dec(?:ember)?)', str(x))))
data
```

```python
# Plotting number of reviews in each month
plt.figure(figsize = (24, 5))
month =['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
sns.despine()
sns.countplot(x="Month of stay", data=data, order=month, color='#62DCEC');
```

```python
data1 = pd.read_csv('D:\LamDong_cmt_Viet1.csv')
data1
```

```python
data1["Year of stay"] = data1["HDate"].apply(lambda x: "".join(re.findall("\d\d\d\d", str(x))))
data1
data1["Year of stay"].unique()
```

```python
# Plotting number of reviews for each year
plt.figure(figsize = (24, 5))
years = ['', '2018', '2017', '2022', '2020', '2016', '2014', '2019', '2015',
       '2021', '2013']
sns.despine()
sns.countplot(x="Year of stay", data=data1, order=years, color='#FB8E7E');
```
